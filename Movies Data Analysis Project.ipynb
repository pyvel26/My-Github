```python
# Using factorize - this assigns a random numeric value for each unique categorical value
```

```python
df1.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
# Assign the factorized output to a new variable and generate a heat map with annotations
```

```python
correlation_matrix = df1.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Take correlation_mat and unstack the variable as corr_pairs.
# Pairs the data with its corresponding values
# The unstack function allows me to reshape the correlation matrix for a better view
```

```python
correlation_mat = df1.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
# Sort corr_pairs with "quicksort" then assign the results to a new variable "sorted_pairs"
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# Now I will take a look at the correlation pairs with a high correlation (> 0.5)
```

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
#I will loop through the dataframe and converted all object data types to category
# Converts all text into numerical values. For example, All country values of 47 represent United States
```

```python

df1_converted = df1


for col_name in df1_converted.columns:
    if(df1_converted[col_name].dtype == 'object'):
        df1_converted[col_name]= df1_converted[col_name].astype('category')
        df1_converted[col_name] = df1_converted[col_name].cat.codes
        
df1_converted
```

```python
# Next I will convert the pearson correlation matrix into a heat map
```

```python
df1_converted.corr(method='pearson')
```

```python
correlation_matrix = df1_converted.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
